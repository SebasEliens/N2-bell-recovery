# Computation times and negativity

```julia
pgfplots(size = (700,233))
times(n,j) = readdlm("../Data/times/timeit_22n_Bell_line_with_kink_n$(n)_c$(["0.00","0.50","1.00"][j]).txt",',')[:,2]
neg(n,j) = readdlm("../Data/times/timeit_22n_Bell_line_with_kink_n$(n)_c$(["0.00","0.50","1.00"][j]).txt",',')[:,1]
t_ = range(0.0, stop = 1.0, length = length(times(2,1)))

p0 = plot(xlabel = "\$\\tau\$", ylabel = "neg", legend = nothing, c=[ ], title = "\\small(a)")
plot!(t_, [neg(2,1) neg(2,2) neg(2,3)], c=[:blue :red :green])



# Initialize array of plots
p = [p0, p0, p0, p0]
for j in 1:3
    clr = [:blue, :red, :green][j]
    p[j+1] = plot(xlabel = "\$\\tau\$", ylabel = "sec", ylims = (0,200), title = j==1 ? "\\small(b)" : "")
    for n in 2:2:12
        plot!(t_, times(n,j), label ="", c = clr)
    end
end

display("Max computation times:")
for j in 1:3
    c = [0.0, 0.5, 1.0]
    display("c=$c -> tmax=$(maximum(times(12,j)))")
end

plt = plot(p..., layout = (1,4))
display(plt)
savefig("pgflot_path_dependent_times.pdf")
```

```julia
pgf_axis = plt.o # pgf axis produced by Plots
PGFPlots.save("pgf/pgflot_path_dependent_times.tex", pgf_axis, include_preamble=false)
```

# Comparison NESTA and LP

```julia
t_nesta = readdlm("../Data/Comparison/nesta/times_22_d.txt",',')
t_lp = readdlm("../Data/Comparison/lp/new_NL_time_100.txt",' ')
```

```julia
using LsqFit
function my_expfit(xs, ys; p0 = [0.5, 1.5], name = "")
    @. model(x, p) = p[1]*(exp(x*p[2])  - 1.0)
    p = coef(curve_fit(model, xs, ys, p0, lower = [0.0,0.0], upper = [Inf,Inf]))
    print("Fit $(name) as a*e^(bx) with (a,b) = $((p[1],p[2]))\n")
    return model(xs ,p)
end

function my_polyfit(xs, ys; p0 = [1.0, 4.0], name = "")
    @. model(x, p) = p[1]*x^p[2]
    p = coef(curve_fit(model, xs, ys, p0, lower = [0.0,0.0], upper = [Inf,Inf]))
       print("Fit $(name) as a*x^b with (a,b) = $((p[1],p[2]))\n")
    return model(xs, p)
end

my_fit = my_polyfit
```

```julia

pgfplots(size=(350,210))
nvals = t_nesta[:,1]
blue = RGBA(85/255, 182/255,255/255, 65/100)
red = RGBA(131/255, 0/255,0/255,39/100)
P = plot(xticks = range(1, stop = 20, step =1), xlabel = "\$n\$", ylabel = "\$10^5\\, \\mathrm{sec}\$") 
plot!(t_lp[:,1],my_fit(t_lp[:,1], t_lp[:,2], name = "LP")/10000, color = red, label = "")
plot!(t_nesta[:,1],my_fit(t_nesta[:,1],t_nesta[:,2], name = "NESTA all")/10000, color = blue, label = "", linestyle = :dash)
plot!(t_nesta[:,1][1:2:end],my_fit(t_nesta[1:2:end,1], t_nesta[1:2:end,2], name = "NESTA odd")/10000, color = blue, label = "")
plot!(t_nesta[:,1][2:2:end],my_fit(t_nesta[2:2:end,1], t_nesta[2:2:end,2], name = "NESTA even")/10000, color = blue, label = "")
plot!(t_nesta[:,1],t_nesta[:,2]/10000,markershape = :diamond, markersize = 9.0, label = "NESTA", line = nothing, color = blue)
plt = plot!(t_lp[:,1],t_lp[:,2]/10000, markershape = :circle, markersize = 7.0, label = "LP", line = nothing, color = red)

savefig("pgf/pgfplot_compare_NEGvsNL.pdf")
display(plt)
```

```julia
pgf_axis = plt.o # pgf axis produced by Plots
PGFPlots.save("pgf/pgfplot_compare_times.tex", pgf_axis, include_preamble=false)
```

# Comparison NEG vs NL

```julia

function compare(d)
    l = readdlm("../Data/Comparison/lp/new_NL22$(d)_100.txt",' ')[:,1];
    n = readdlm("../Data/Comparison/nesta/neg_22_$(d).txt",',')[:,1];
    p = sortperm(l);
    return l[p], n[p]
end
```

```julia
pgfplots(size=(350,210))
plt1 =  plot(xlabel = "NL", ylabel = "neg")
for d in 2:2:12
    plot!(compare(d)..., label = "$d",markershape = :circle)
end

plt2 =  plot(xlabel = "NL", ylabel = "neg")
for d in 3:2:12
    plot!(compare(d)..., label = "$d",markershape = :circle)
end

plt = plot(plt1,plt2,layout =2)
```
